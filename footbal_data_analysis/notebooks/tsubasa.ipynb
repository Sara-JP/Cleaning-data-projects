<h2 align=right style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazir" color="#0099cc">
قسمت اول
</font>
</h2>

<p dir=rtl style="direction: rtl;text-align: right;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
    به عنوان اولین قسمت از حل سوال، میخواهیم ستونی به اسم <code>label</code> به <code>df</code> اضافه کنیم.
    <br>
    در این سوال ما <b>گُل</b> و <b>گُل به خودی</b> را گل در نظر می‌گیریم. به عبارت بهتر، در پیشبینی احتمال گل، آن‌ها را معادل یکدیگر لحاظ می‌کنیم.
    <br>
    بنابراین ستون <code>label</code> به ازای سطرهایی که ستون <code>outcome</code>شان برابر گُل یا گُل به‌خودی است مقدار ۱ و در غیر این صورت مقدار ۰ خواهد داشت.
    <br>
    پس از محاسبه ستون <code>label</code> ستون <code>outcome</code> را برای همیشه از <code>df</code> حذف کنید.
</font>
</p>

```python
#TODO: add `label` column and remove `outcome` from df
df['label'] = df['outcome'].apply(lambda x: 1 if x in ["گُل", "گُل به خودی"] else 0)
df = df.drop('outcome', axis=1)
df
```


<p dir=rtl style="direction: rtl;text-align: right;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
    برای آن که ارزش کار خود را درک کنید، نیاز است ابتدا که هیچگونه پیش‌پردازشی روی داده انجام نشده است، یکبار با آن مدلسازی کنید و عملکرد مدل را مشاهده کنید. با اجرای سلول زیر یک مدل روی داده آموزش می‌بیند و نتایجی تولید می‌کند.
    <br>
    اگر اجرای این سلول با شکست مواجه شد، می‌تواند به یکی از دلایل زیر باشد.
</font>
</p>

            کتابخانه Fast Auto ML را نصب نکرده باشید.

            در بارگزاری مجموعه داده دچار مشکل شده باشید.

```python
from sklearn.model_selection import train_test_split
from flaml import AutoML
from sklearn.metrics import roc_auc_score

x_train,x_test , y_train,y_test = train_test_split(df.drop('label', axis=1), df.label, random_state=313, stratify=df.label)

model = AutoML(task='classification', time_budget=60, verbose=0)
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
print(f'performance of model is {roc_auc_score(y_test, y_pred)}')
```

به عنوان گام سوم، قصد داریم پای راست و چپ را معادل یکدیگر در نظر بگیریم.
    <br>
    بنابراین در قسمت سوم، مقادیر <b>پای راست</b> و <b>پای چپ</b> در ستون <code>bodyPart</code> را با کلمه <b>پا</b> جایگزین کنید.

<h2 align=right style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazir" color="#0099cc">
قسمت چهارم
</font>
</h2>

<p dir=rtl style="direction: rtl;text-align: right;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
    در قسمت چهارم قصد داریم، فاصله مبدا شوت و زاویه آن را به عنوان دو ستون جدید با نام‌های <code>distance</code> و <code>angle</code> به مجموعه داده اضافه کنیم.
    <br>
    در ادامه فرمول‌هایی که طبق آن‌ها می‌توانیم فاصله و زاویه را محاسبه کنیم، آورده شده است.
    <br>
</font>
</p>


<p dir=rtl style="direction: rtl;text-align: center;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
    $distance = \sqrt{x^2 + y^2}$
    <br>
    <br>
    
$$ angle=  \left\{
\begin{array}{}
      rad2deg(arctan(\theta)) & arctan(\theta) \geq 0  \\
      rad2deg(arctan(\theta) + \pi) & arctan(\theta) \lt 0 \\
\end{array} 
\right. , \theta = \frac{7.32 * x}{x^2 + y^2 - (7.32 /2)^2} $$ 
    <br>
</font>
</p>



<p dir=rtl style="direction: rtl;text-align: right;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
    از دو فرمول بالا می‌توانیم نتیجه بگیریم، ستون <code>distance</code> باید از جنس float باشد.
    <br>
    همچنین متوجه می‌شویم که اعداد ستون <code>angle</code> که از جنس float هستند، بر حسب <b>درجه</b> حساب شده اند.
    <br>
    برای محاسبه عدد پی از np.pi استفاده کنید.
    <br>
    پس از محاسبه <code>distance</code> و <code>angle</code> لازم است دو ستون <code>x</code> و <code>y</code> را برای همیشه از دیتافریم حذف کنید. 
</font>
</p>



<p dir=rtl style="direction: rtl;text-align: right;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
    به عنوان تمرین اضافه، تلاش کنید درستی فرمولی که برای محاسبه زاویه ارائه شد را اثبات کنید.
    <br>
    راهنمایی : ۷.۳۲، طول دروازه است!
    <br>
    اثبات این فرمول، هیچ تاثیری در امتیاز کسب شده توسط شما برای این سوال نخواهد داشت و صرفا برای تمرین بیشتر شما طرح شده است.
</font>
</p>


```python
#TODO: add distance and angle column to df
df['distance'] = (df['x']**2 + df['y']**2)**0.5
tetha= (7.32*df['x'])/(df['x']**2 + df['y']**2-(7.32/2)**2)
df['angle']= tetha.apply(lambda x: np.rad2deg(np.arctan(x)) if np.arctan(x)>=0 else np.rad2deg(np.arctan(x)+np.pi))
df = df.drop('x', axis=1)
df = df.drop('y', axis=1)
df
```

<p dir=rtl style="direction: rtl;text-align: right;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
    با اجرای سلول زیر، متوجه خواهیم شد که در ستون <code>interferenceOnShooter</code> تعدادی مقدار گمشده وجود دارد که پیش از مدلسازی نهایی، لازم است آن را جبران کنیم.
    <br>
    همانطور که در توضیحات مجموعه داده گفته شد، ستون <code>interveningOpponents</code> نشان دهنده افرادی از تیم مقابل است که مانع دید شوت‌زننده بوده‌اند. از طرفی ستون <code>interferenceOnShooter</code> هم میزان اثرگذاری بازیکننان تیم مقابل را نشان می‌دهد. بنابراین ایده خوبی است که برای پر کردن مقادیر گمشده آن، از <code>interveningOpponents</code> استفاده کنیم.
    <br>
    بنابراین برای پرکردن مقادیر گمشده ستون <code>interferenceOnShooter</code>در هر سطر، اگر ستون <code>interveningOpponents</code> آن ۰ بود، عبارت <b>کم</b>، اگر برابر ۱ بود، عبارت <b>متوسط</b> و در غیر این صورت عبارت <b>زیاد</b> را جایگزین کنید.
</font>
</p>

```python
#TODOL calculate feature importance and initialize featuer_importance dataframe

from sklearn.feature_selection import mutual_info_classif
feature_importance = pd.DataFrame()

X = df.drop('label', axis=1)
y = df['label']

feature_importance = mutual_info_classif(X, y, random_state=1401)

feature_importance_df = pd.DataFrame(data=feature_importance, index=X.columns, columns=['fi'])

feature_importance_df = feature_importance_df.sort_values(by='fi', ascending=False)

feature_importance=feature_importance_df
feature_importance
```
